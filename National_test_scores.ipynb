```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#Import All CSV Files
#math
math_09 = pd.read_csv("Resources/Education Data/2009 Math .csv")
math_09_df = pd.DataFrame(math_09)
math_11 = pd.read_csv("Resources/Education Data/2011 Math .csv")
math_11_df = pd.DataFrame(math_11)
math_13 = pd.read_csv("Resources/Education Data/2013 math.csv")
math_13_df = pd.DataFrame(math_13)
math_15 = pd.read_csv("Resources/Education Data/2015 Math .csv")
math_15_df = pd.DataFrame(math_15)
math_17 = pd.read_csv("Resources/Education Data/2017 Math .csv")
math_17_df = pd.DataFrame(math_17)
math_19 = pd.read_csv("Resources/Education Data/2019 Math.csv")
math_19_df = pd.DataFrame(math_19)

#reading
reading_09 = pd.read_csv("Resources/Education Data/2009 Reading .csv")
reading_09_df = pd.DataFrame(reading_09)
reading_11 = pd.read_csv("Resources/Education Data/2011 reading .csv")
reading_11_df = pd.DataFrame(reading_11)
reading_13 = pd.read_csv("Resources/Education Data/2013 reading.csv")
reading_13_df = pd.DataFrame(reading_13)
reading_15 = pd.read_csv("Resources/Education Data/2015 Math .csv")
reading_15_df = pd.DataFrame(reading_15)
reading_17 = pd.read_csv("Resources/Education Data/2017 Reading .csv")
reading_17_df = pd.DataFrame(reading_17)
reading_19 = pd.read_csv("Resources/Education Data/2019 Reading.csv")
reading_19_df = pd.DataFrame(reading_19)
```

```python
# Preview dfs
math_09_df
```

```python
#Drop SigDiff and SigSymbol off all csv
#math
math_09_df = math_09_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
math_11_df = math_11_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
math_13_df = math_13_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
math_15_df = math_15_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
math_17_df = math_17_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
math_19_df = math_19_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)

#reading
reading_09_df = reading_09_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
reading_11_df = reading_11_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
reading_13_df = reading_13_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
reading_15_df = reading_15_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
reading_17_df = reading_17_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
reading_19_df = reading_19_df.drop(['SigDiff', 'SigSymbol', 'AB', 'AP'], axis=1)
```

```python
#preview dfs
reading_09_df
```

```python
#rename column to include year, change Jurisdiction to State
#math
math_09_df = math_09_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Math Score 09'})
math_11_df = math_11_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Math Score 11'})
math_13_df = math_13_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Math Score 13'})
math_15_df = math_15_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Math Score 15'})
math_17_df = math_17_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Math Score 17'})
math_19_df = math_19_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Math Score 19'})

#reading
reading_09_df = reading_09_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Reading Score 09'})
reading_11_df = reading_11_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Reading Score 11'})
reading_13_df = reading_13_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Reading Score 13'})
reading_15_df = reading_15_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Reading Score 15'})
reading_17_df = reading_17_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Reading Score 17'})
reading_19_df = reading_19_df.rename(columns = {'Jurisdiction' : 'State', 'MN' : 'NAEP Reading Score 19'})

```

```python
#preview dfs

reading_15_df
```

```python
#Merge the dfs
#math
new_math_df = math_09_df.merge(math_11_df, on = 'State')
new_math_df = new_math_df.merge(math_13_df, on = 'State')
new_math_df = new_math_df.merge(math_15_df, on = 'State')
new_math_df = new_math_df.merge(math_17_df, on = 'State')
new_math_df = new_math_df.merge(math_19_df, on = 'State')

#reading
new_reading_df = reading_09_df.merge(reading_11_df, on = 'State')
new_reading_df = new_reading_df.merge(reading_13_df, on = 'State')
new_reading_df = new_reading_df.merge(reading_15_df, on = 'State')
new_reading_df = new_reading_df.merge(reading_17_df, on = 'State')
new_reading_df = new_reading_df.merge(reading_19_df, on = 'State')
```

```python
#preview dfs
new_math_df
```

```python
#reset state as index
reset_math_df = new_math_df.set_index('State')
reset_reading_df = new_reading_df.set_index('State')

#drop DoDEA and Puerto Rico
clean_math_df = reset_math_df.drop(['DoDEA', 'Puerto Rico'])
clean_reading_df = reset_reading_df.drop(['DoDEA', 'Puerto Rico'])
```

```python
#preview dfs
clean_math_df
```

```python
#calc average over time
clean_math_df['Average NAEP Math Score over 10 Years'] = clean_math_df.mean(axis = 1)
final_math_df = clean_math_df['Average NAEP Math Score over 10 Years']
final_math_df = final_math_df.round(2)
final_math_df = final_math_df.to_frame()
final_math_df = final_math_df.reindex(['National public', 'Massachusetts', 'New Hampshire', 'Minnesota', 'Vermont', 'New Jersey',
       'Kansas', 'North Dakota', 'Connecticut', 'Maine', 'Montana', 'Maryland',
       'North Carolina', 'Ohio', 'Pennsylvania', 'Wisconsin', 'Colorado',
       'Virginia', 'Indiana', 'Iowa', 'Washington', 'South Dakota', 'Wyoming',
       'Florida', 'Idaho', 'Missouri', 'New York', 'Texas', 'Utah', 'Delaware', 'Kentucky', 'Rhode Island', 'Nebraska', 'Illinois',
       'Oregon', 'Arkansas', 'Alaska', 'Oklahoma', 'Michigan', 'Georgia',
       'Hawaii', 'South Carolina', 'Nevada', 'West Virginia', 'Tennessee',
       'California', 'New Mexico', 'Arizona', 'Louisiana', 'Alabama',
       'Mississippi', 'District of Columbia'])

clean_reading_df['Average NAEP Reading Score over 10 Years'] = clean_reading_df.mean(axis = 1)
final_reading_df = clean_reading_df['Average NAEP Reading Score over 10 Years']
final_reading_df = final_reading_df.round(2)
final_reading_df = final_reading_df.to_frame()
final_reading_df = final_reading_df.reindex([ 'National public','Massachusetts', 'New Jersey', 'New Hampshire', 'Connecticut',
       'Vermont', 'Virginia', 'Maryland', 'North Dakota', 'Colorado',
       'Florida', 'Kentucky', 'Delaware', 'Montana', 'Ohio', 'New York',
       'Kansas', 'Missouri', 'Maine', 'Pennsylvania', 'Minnesota',
       'Rhode Island', 'Indiana', 'Wyoming', 'Nebraska', 'South Dakota',
       'Iowa', 'Washington', 'Idaho', 'Wisconsin',
       'North Carolina', 'Utah', 'Illinois', 'Texas', 'Michigan', 'Oregon',
       'Georgia', 'Oklahoma', 'Tennessee', 'Alabama', 'Arkansas',
       'South Carolina', 'West Virginia', 'Nevada', 'Alaska', 'Hawaii',
       'Mississippi', 'Arizona', 'California', 'New Mexico', 'Louisiana',
       'District of Columbia'])
```

```python
#identify the 5 highest average scores
highest_math_df = final_math_df.sort_values(['Average NAEP Math Score over 10 Years'], ascending=False)
top_5_math_df = highest_math_df.head(5)
print(top_5_math_df)

highest_reading_df = final_reading_df.sort_values(['Average NAEP Reading Score over 10 Years'], ascending=False)
top_5_reading_df = highest_reading_df.head(5)
print(top_5_reading_df)
```

```python
#identify the 5 lowest average scores
lowest_math_df = final_math_df.sort_values(['Average NAEP Math Score over 10 Years'], ascending=True)
bottom_5_math_df = lowest_math_df.head(5)
print(bottom_5_math_df)

lowest_reading_df = final_reading_df.sort_values(['Average NAEP Reading Score over 10 Years'], ascending=True)
bottom_5_reading_df = lowest_reading_df.head(5)
print(bottom_5_reading_df)
```
